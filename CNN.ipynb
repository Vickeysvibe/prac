```python
import tensorflow as tf
from tensorflow import keras
from keras import Model
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```python
training_set = train_datagen.flow_from_directory('F:/dataset/train',
                                                 target_size = (64,64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```python
test_set=test_datagen.flow_from_directory('F:/dataset/validation',target_size=(64,64),batch_size=32,class_mode='binary')
```

```python
from keras.layers import Conv2D,MaxPool2D,Dense,Flatten
from keras.models import Model,Sequential
```

```python
cnn=Sequential()
```

```python
cnn.add(Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))
```

```python
cnn.add(MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(Conv2D(filters=32,kernel_size=3,activation='relu'))
```

```python
cnn.add(MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(Flatten())
```

```python
cnn.add(Dense(units=128,activation='relu'))
cnn.add(Dense(units=1,activation='sigmoid'))
```

```python
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(x=training_set,validation_data=test_set,epochs=50)
```

```python
import numpy as np
import keras.utils as image
test_image = image.load_img('F:/downloaddog.jpeg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'
print(prediction)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import numpy as np

# Assume X_train is a batch of dog/cat images, y_train is their labels (0: cat, 1: dog)
X_train = np.random.rand(100, 64, 64, 3)  # Sample data
y_train = np.random.randint(0, 2, 100)  # Sample labels

# CNN Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10)

# Predict
predictions = model.predict(X_train[:5])
print("Predictions:", predictions)
```
